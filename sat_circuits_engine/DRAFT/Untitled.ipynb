```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

qc = QuantumCircuit(3)
qc.x([0, 1])
qc.x(1)
qc.barrier()
qc.rccx(0 ,1 ,2)
qc.barrier()
qc.ccx(0, 1, 2)
qc.measure_all()

display(qc.draw('mpl'))
display(qc.decompose().draw('mpl'))

job = AerSimulator().run(transpile(qc, AerSimulator()))
results = job.result()
counts = results.get_counts()

print(counts)
```
